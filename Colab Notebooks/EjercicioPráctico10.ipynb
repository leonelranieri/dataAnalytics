#Actividad 2: Análisis de Supervivencia
Cargá la base de datos “titanic” y visualizá las primeras filas.

Realizá un análisis estadístico básico de la tasa de supervivencia por clase de pasajero.

Creá un gráfico de barras que muestre la tasa de supervivencia por clase.

Analizá si existe alguna correlación entre la edad de los pasajeros y su tasa de supervivencia, utilizando un diagrama de caja (box plot).


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
```

```python
titanic.info()
```

```python
# Agrupar por clase y calcular la media de la columna 'survived'
tasa_supervivencia_clase = titanic.groupby("pclass")["survived"].mean().reset_index()

# Renombrar la columna para mayor claridad
tasa_supervivencia_clase.columns = ["Clase", "Tasa de Supervivencia"]

# Mostrar resultados
tasa_supervivencia_clase
```

```python
# cut() permite agrupar valores numéricos continuos y clasificarlos en intervalos
"""x: la serie o lista de números que querés clasificar.
bins: los límites de los intervalos (lista de números).
labels (opcional): etiquetas para cada intervalo. Si no se ponen, se muestran los rangos como (a, b].
right: si True (por defecto), el límite derecho del intervalo está incluido. Ej: (a, b].
include_lowest: si True, incluye el límite más bajo en el primer intervalo. """

# Definir rangos de edad simples
titanic["Edad_Rango"] = pd.cut(titanic["age"], bins=[0, 18, 30, 40, 50, 60, 80],
                               labels=["0-18", "19-30", "31-40", "41-50", "51-60", "61-80"])

# Agrupar por rango de edad y calcular la tasa de supervivencia
tasa_supervivencia_edad = titanic.groupby("Edad_Rango")["survived"].mean().reset_index()

# Renombrar la columna para mayor claridad
tasa_supervivencia_edad.columns = ["Rango-Edad", "Tasa de Supervivencia"]

print(tasa_supervivencia_edad)
```

```python
# Crear gráfico de barras
plt.bar(tasa_supervivencia_clase['Clase'].astype(str), # convertir a string para que matplotlib trate como categoría
        tasa_supervivencia_clase['Tasa de Supervivencia'], color='skyblue')
plt.title('Tasa de Supervivencia por clase')
plt.xlabel('Clase')
plt.ylabel('Tasa de supervivencia')
plt.show()
```

```python
sns.boxplot(data=tasa_supervivencia_edad, x='Rango-Edad', y='Tasa de Supervivencia')
plt.title('Tasa de Supervivencia por Rango de Edad')
plt.xlabel('Rango de Edad')
plt.ylabel('Tasa de supervivencia')
plt.show()
```

El gráfico muestra que el rango etario '31-40' alcanza un nivel esperado de supervivencia lo mismo que el primer rango etario. Uno por, se supone, propios medios y el otro por considerarse prioritario en rescates. Tal vez sorprende que el rango etario '51-60' hata superado a su antecesor. Tal vez este asociado con el tipo de clase en la que viajaba.
