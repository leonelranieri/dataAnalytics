# ACTIVIDAD 1 :

Cargar el conjunto de datos , que contiene informaci贸n sobre las ventas de diferentes productos en distintos meses.

Agrupar las ventas por producto y sumar las ventas totales para cada uno.

Generar un nuevo DataFrame que contenga el total de ventas por producto y lo mostrarlo en pantalla

```python
import pandas as pd

data = {
'producto': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],
'mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo', 'Marzo', 'Enero', 'Febrero'],
'ventas': [150, 200, 250, 300, 100, 400, 350, 200, 300]
}

# Cargar datos
df = pd.DataFrame(data)

# Agrupar ventas por producto y sumar ventas totales para cada uno
ventas_por_producto = df.groupby('producto')['ventas'].sum()
#print(ventas_por_producto)

# Nuevo DataFrame que contenga ventas por producto
df_ventas_por_producto = df.groupby('producto')['ventas'].sum().reset_index(name='Ventas por producto')

print(df_ventas_por_producto)
```

Aplicar la funci贸n pivot_table() de Pandas para resumir informaci贸n clave de los datos de ventas.

Crear un resumen que permita analizar las ventas mensuales por producto.

```python
import pandas as pd

data = {
'producto': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],
'mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo', 'Marzo', 'Enero', 'Febrero'],
'ventas': [150, 200, 250, 300, 100, 400, 350, 200, 300]
}

df = pd.DataFrame(data)

df
```

```python
df_ventas_mensuales_por_producto = df.pivot_table(index='producto', columns='mes', values='ventas', aggfunc='sum')

df_ventas_mensuales_por_producto

```
