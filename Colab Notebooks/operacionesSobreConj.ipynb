```python
A = {1, 2, 3, 4}
B = {2, 4, 6, 8}

# unión : todos los elementos de A y B
C = A | B
print(c)

# intersección : elementos comunes de A y B
D = A & B
print(D)

# diferencia : elementos de A que no están en B
E = A - B
print(E)

# diferencia simétrica : elementos que están en A o en B, pero no en ambos
F = A ^ B
print(F)
```
