```python
import pandas as pd
import numpy as np
from scipy import stats

ventas = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vS52Bmhv9F9gN6c4xb1pACABjA0gfzm9B9D9-Gem65S6hY6d3eXGOl_VsyhwsLoer6AGnkIqKpQYVK1/pub?output=csv')
ventas['precio'] = ventas['precio'].str.replace('$', '')
ventas['precio'] = ventas['precio'].astype(float)

ventas = ventas.dropna(subset=['precio'])

mediaDeVentas = np.mean(ventas['precio']).round(2)
print("media de ventas : ", mediaDeVentas)

medianaDeVentas = np.median(ventas['precio']).round(2)
print("mediana de ventas : ", medianaDeVentas)

modaVentas = stats.mode(ventas['precio'])[0]
print("moda de ventas : ", modaVentas)

rangoVentas = np.max(ventas['precio']) - np.min(ventas['precio'])
print("rango de ventas : ", rangoVentas)

varianzaVentas = round(np.var(ventas['precio']), 2)
print("varianza de ventas : ", varianzaVentas)

desviacionVentas = round(np.std(ventas['precio']), 2)
print("desviacion de ventas : ", desviacionVentas)

```
