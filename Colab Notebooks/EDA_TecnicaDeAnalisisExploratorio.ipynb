## introducción a la visualización de datos
permite representar información de manera gráfica.
Matplotlib y Seaborn permiten crear variedad de gráficos.


```python
# Visualización con Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Crear dataframe
data = {
    'producto' : ['A', 'B', 'C', 'D'],
    'ventas' : [150, 200, 300, 100]
}

df = pd.DataFrame(data)

# Crear gráfico de barras
plt.bar(df['producto'], df['ventas'], color='skyblue')
plt.title('Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.show()
```

```python
# Visualización con Seaborn
import seaborn as sns

# Crear gráfico de distribución de ventas
sns.histplot(df['ventas'], bins=4, kde=True)
plt.title('Distribución de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()
```

## Ejemplo con base de datos 'tips'

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset('tips')
tips.describe()

# el total de las cuentas varía entre 3.07 y 50.81 con un promedio de 19.78
# lo mismo se puede expresar para tip y size.
```

## análisis de distribución
utilizando histogramas

```python
# Crea una ventana de figura
plt.figure(figsize=(12, 5))

# Visualizar la distribución de las propinas
plt.subplot(1, 2, 1)
sns.histplot(tips['tip'], bins=30, kde=True)
plt.title('Distribución de Propinas')
plt.xlabel('Propinas')
plt.ylabel('Frecuencia')

# Visualizar la distribución del total de las cuentas
plt.subplot(1, 2, 2)
sns.histplot(tips['total_bill'], bins=30, kde=True)
plt.title('Distribución del total de las cuentas')
plt.xlabel('Total de la cuenta')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# subplot crea un subgráfico dentro de la figura (fila, columna, pos)
# (2,1,1) y (2,1,2) crea dos gráficos peero uno arriba y otro abajo.
```

## Visualización con Boxplot

```python
# crear una ventana de figura
plt.figure(figsize=(12, 5))

# Visualizar la distribución de las propinas
plt.subplot(1, 2, 1)
sns.boxplot(tips['tip'])
plt.title('Distribución de Propinas')

# Visualizar la distribución de los totales de la cuentas
plt.subplot(1, 2, 2)
sns.boxplot(tips['total_bill'])
plt.title('Distribución de Totales de Cuentas')

tips['total_bill'].describe()
```

## Análisis de la relación entre variables
uso de diagrama de dispersión.

```python
plt.figure(figsize=(8, 6))

# Visualizar la relación entre el total de la cuenta y la propina
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='sex', s=100)
plt.title('Relación entre Total de la Cuenta y Propinas')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propinas')
plt.show()

# a medida que aumenta la cuenta aumenta la propina
# propinas más altas pagadas por hombres los sábados y domingos
```
