#FILTROS Y SELECCIONES

```python
import pandas as pd


data = {
    'Producto' : ['A', 'B', 'C', 'D', 'E'],
    'Ventas' : [120, 135, 118, 97, 148],
    'Fecha' : ['01/01/2022', '05/04/2022', '03/05/2022', '18/09/2022', '24/08/2022']
}

df = pd.DataFrame(data)

#Info del dataframe
#df.info()

#Nombres de columnas
columnas = df.columns
print(columnas)
print()
print()
#Índices del dataframe
indices = df.index
print(indices)
print()
print()
#Obtener datos de una columna
ventas = df['Ventas']
print('Datos de la columna ventas:\n', ventas)
print()
print()
#Obtener datos de una fila
fila_1 = df.loc[1]
print('Datos de la fila 1:\n', fila_1)
print()
print()
# Seleccionar varias columnas
sub_df = df[['Ventas', 'Producto']]
print('Datos de las columnas Ventas y Producto:\n', sub_df)
print()
print()
# Seleccionar filas y columnas
sub_set = df.loc[1:3, ['Producto', 'Fecha']]
print('Datos de las filas 1, 2 y 3 de las columnas Producto y Fecha:\n', sub_set)
print()
print()
# Selección condicional: ventas mayores a 90 y menores a 125
condicion = (df['Ventas'] > 90) & (df['Ventas'] < 125)
print('Ventas mayores a 90 y menores a 125:\n', df[condicion])
print()
print()
# Búsqueda condicional utilizando 'query'
nueva_condicion = df.query('Ventas > 90 and Ventas < 125')
print('Ventas mayores a 90 y menores a 125:\n', nueva_condicion)
print()
print()
#Crear una nueva columna con el 10$ de aumento
df['Ventas con aumento'] = df['Ventas'] * 1.1
print('Ventas con aumento:\n', df)
print()
print()
# Eliminar columna
df = df.drop('Ventas con aumento', axis=1) # opción sin axis df = df.drop(columns='Ventas con aumento')
print('Ventas sin aumento:\n', df)
print()
print()
# Eliminar una fila
df = df.drop(2)
print('DF sin la fila 2:\n', df)
print()
print()
# Reindexar el df
df = df.reset_index(drop=True)
print('DF con reindexado:\n', df)
print()
print()
# Transponer el df
df_transpuesto = df.T
print('DF transpuesto:\n', df_transpuesto)
print()
print()
```
