# carga de los dataframes

```python
import pandas as pd

df_sur = pd.read_excel("https://docs.google.com/spreadsheets/d/e/2PACX-1vSqWIQIAjT3cNemFoxQ4VpvkNvPDBqXowhf1TcxIjPdvXR00_gvm_qDiS5rOnfzPg/pub?output=xlsx", sheet_name="Sur")
df_norte = pd.read_excel("https://docs.google.com/spreadsheets/d/e/2PACX-1vSqWIQIAjT3cNemFoxQ4VpvkNvPDBqXowhf1TcxIjPdvXR00_gvm_qDiS5rOnfzPg/pub?output=xlsx", sheet_name="Norte")

print(df_sur.head())
print(df_norte.head())
```

# exploración de datos

```python
# determinar si los DF contienen los mismos valores en la columna 'Producto'

productos_sur = df_sur['Producto'].unique()
productos_norte = df_norte['Producto'].unique()

set_sur = set(productos_sur)
set_norte = set(productos_norte)

print(set_sur == set_norte)

```

```python
# controlar cantidad de recurrencias por producto

print(df_sur['Producto'].value_counts())
print(df_norte['Producto'].value_counts())

```

```python
#asegurar la misma estructuras en las columnas
# modificar los nombres de las columnas de df_norte para que coincidan con las de df_sur

df_norte_renombrado = df_norte.rename(columns={
    "Sucursal": "Sede",
    "Ingresos": "Ventas"
})

df_norte_renombrado
```

#concatenar los DataFrames

```python
# unir verticalmente --> pd.concat()

df_final = pd.concat([df_sur, df_norte_renombrado], ignore_index=True)
df_final
```

Utiliza pd.merge() para combinar ambos DataFrames basados en la columna que se considere más apropiada, con el join que corresponda. Dejar documentada la justificación de ambas elecciones.

```python
import pandas as pd

# obtención de los dataframes

df_data_empleado = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQRroWZdUymNhEZzbIvp5uxBP6h7nDlRdRXWMrGvJmMQ_scjNSKjOt-HTGUK5JJqRYrr0LSbe2sRpI7/pub?output=csv")
df_situacion_empleado = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQh8NpLeggg7UN1pBuBW-Rp6jiEm2BDwHseLnjhlZ1zZhSX5fyLzHZFt4qZL0J7c2urbeP_ZyvChMpJ/pub?output=csv")

print(df_data_empleado.info())
print()
print(df_situacion_empleado.info())
```

```python
 # aplicación de la función merge

 df_merge = pd.merge(df_data_empleado, df_situacion_empleado, how='outer', on='ID', sort=True, indicator=True)
 df_merge

# registros únicos
 df_left_only = df_merge[df_merge['_merge'] == 'left_only']
 df_left_only.info()

 df_right_only = df_merge[df_merge['_merge'] == 'right_only']
 df_right_only.info()

 # ids sin coincidencia
 id_df1 = set(df_data_empleado['ID'])
 id_df2 = set(df_situacion_empleado['ID'])

 solo_en_df1 = id_df1 - id_df2
 solo_en_df2 = id_df2 - id_df1

 print(solo_en_df1)
 print(solo_en_df2)
```
