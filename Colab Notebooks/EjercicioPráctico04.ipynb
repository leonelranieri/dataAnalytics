```python
import pandas as pd

df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSGx7xckqyPgWaVaMqhqHascxmBjT4rto6VFjLBE7t5pdqHbcOoAWWXnyEYGKzThIb3PSQdEtHxHI_y/pub?output=csv')

total_registros_shape = df.shape[0] #devuelve una tupla(fila, columna)
total_registros_len = len(df)

print(total_registros_shape)
print(total_registros_len)
```

```python
  #valores nulos por columna
  df.isnull().sum()
```

```python
# cantidad total de filas duplicadas
filas_duplicadas = df.duplicated().sum()

print(filas_duplicadas)
```

NUEVO DATA FRAME SÓLO CON LOS DUPLICADOS

```python
df_duplicados = df[df.duplicated(keep=False)]

print(df_duplicados)  #marca todas las apariciones duplicadas
print(len(df_duplicados))
```

```python
df_duplicados_unicos = df[df.duplicated()].drop_duplicates()  #cuenta duplicados únicos

print(df_duplicados_unicos)
print(len(df_duplicados_unicos))
```
