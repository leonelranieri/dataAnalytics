Actividad 1 : tipos de par치metros.
- posicionales
- por defecto
- indefinidos
- de palabras claves indefinidas

- POSICIONALES

```python
def multiplicar(x,y):
    return x * y

print(multiplicar(1, 1))
print(multiplicar(-2, 8))
print(multiplicar(0, 45))
print(multiplicar(2, 0.5))
```

POR DEFECTO

```python
def bienvenida(nombre="Bienvenida"):
  return f"Hola, {nombre}\n"

print(bienvenida("Juan"))
print(bienvenida())
```

INDEFINIDOS

```python
def calcular_promedio(*args):
  return sum(args) / len(args)

print(calcular_promedio(10,2))
print(calcular_promedio(10,2,3))
print(calcular_promedio(10,2,3,5))
```

DE PALABRAS CLAVES INDEFINIDAS

```python
def concatena_info(nombre, **kwargs):
  name = f"Nombre: {nombre}"
  for key, value in kwargs.items():
    name += f", {key}: {value}"
  return name

print(concatena_info("Juan Soria", Edad=32, DNI=38748568, Ciudad="CABA"))
print(concatena_info("Juana Coria", Edad=35, DNI=35748568, Ciudad="Berazategui"))
print(concatena_info("Xuan Kotikia", Edad=38, DNI=33748568, Ciudad="CABA"))

```
