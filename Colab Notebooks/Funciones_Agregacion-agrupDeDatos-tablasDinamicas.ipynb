```python
 # -------------------- AGREGACIÓN --------------------------------

 #  Resumir varios registros en un único valor.

 import pandas as pd

 data = {
     'Producto' : ['A', 'B', 'A', 'B', 'C'],
     'Ventas' :[200, 150, 300, 200, 400]
 }

 df = pd.DataFrame(data)

 ventas_totales = df['Ventas'].sum()  # EN SQL : SELECT SUM(Ventas) FROM tabla
 print(ventas_totales)

# --------------------- AGRUPAMIENTO DE DATOS ---------------------------
ventas_por_producto = df.groupby('Producto')['Ventas'].sum() # EN SQL : SELECT Producto, SUM(Ventas) FROM Tabla GROUP BY Producto

# Segmenta la información en grupos basados en características específicas.

 print(ventas_por_producto)

 # ------------ TABLAS DINÁMICAS -----------------------

 # Permiten realixar análisis interactivos

 tabla_dinamica = df.pivot_table(index='Producto', values='Ventas', aggfunc='sum')
 tabla_dinamica
```
