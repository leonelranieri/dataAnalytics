```python
import pandas as pd
import seaborn as sns

df_titanic = sns.load_dataset('titanic')
df_titanic
```

```python
df_titanic_alive = df_titanic.query("alive == 'yes'")
df_titanic_noAlive = df_titanic.query("alive == 'no'")
print("pasajeros sobrevivientes: ", df_titanic_alive.shape[0])
print("pasajeros no sobrevivientes: ", df_titanic_noAlive.shape[0])
```

```python
df_titanic_alive.columns
df_filtrado = df_titanic_alive[['sex', 'age', 'fare']]
df_filtrado
```

```python
df_filtrado['categoria'] = df_titanic["class"].map({
    "First" : "A",
    "Second" : "B",
    "Third" : "C"
})
df_filtrado
```
